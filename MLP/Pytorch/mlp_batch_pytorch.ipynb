```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import torch 
import tensorflow as tf
from torch.utils.data import DataLoader, TensorDataset
from torch import Tensor
```

```python

data=pd.read_csv('IRIS.csv')
#data=pd.get_dummies(data,columns=['species'])
df=data
species = dict(zip(list(df['species'].unique()), ([0, 1, 2])))
df['species'].replace(species,inplace=True)

y=df.pop('species')
x=df.values
```

```python
X_torch=torch.tensor(x,dtype=torch.float32)
y_torch=torch.tensor(y,dtype=torch.long)
```

```python
dataset=TensorDataset(Tensor(X_torch),Tensor(y_torch))
loader=DataLoader(dataset,batch_size=16)
```

```python
class MLPTorch(torch.nn.Module):
    def __init__(self,input_dim,units=32,num_class=3):
        super(MLPTorch,self).__init__()
        self.linear1=torch.nn.Linear(input_dim,units)
        self.linear2=torch.nn.Linear(units,64)
        self.linear3=torch.nn.Linear(64,32)
        self.linear4=torch.nn.Linear(32,num_class)
        self.relu=torch.nn.ReLU()
        self.softmax=torch.nn.Softmax()
    def __call__(self,xb):
        l1=self.linear1(xb)
        r1=self.relu(l1)
        l2=self.linear2(r1)
        r2=self.relu(l2)
        l3=self.linear3(r2)
        r3=self.relu(l3)
        l4=self.linear4(r3)
        out=self.softmax(l4)
        return out    
    def calc_accuracy(self,y_pred,y_torch):

        return (torch.argmax(y_pred,dim=1)==y_torch).float().sum()/len(y_torch)    
```

```python
model=MLPTorch(X_torch.shape[1],units=32,num_class=3)
criterion=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
for epoch in range(200):

    total_loss=0
    total_acc=0
    for i,data in enumerate(loader):
        X,Y=data
        y_pred=model(X)
        loss_function=criterion(y_pred,Y)
        optimizer.zero_grad()
        loss_function.backward()
        optimizer.step()
        acc=model.calc_accuracy(y_pred,Y)
        total_loss+=loss_function.detach().numpy()
        total_acc+=acc
       
    total_loss/=i+1
    total_acc/=i+1

    print("after {} epoch, loss ={:.4f} and accuracy is {:.4f}".format(epoch,total_loss,total_acc))
    
```

```python
for epoch in range(2):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(loader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
    #if i % 2000 == 1999:    # print every 2000 mini-batches
    

print('Finished Training')
```
