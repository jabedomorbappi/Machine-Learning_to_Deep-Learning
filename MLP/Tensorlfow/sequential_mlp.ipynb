```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import tensorflow as tf 

from sklearn.preprocessing import StandardScaler

import time
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data=pd.read_csv('IRIS.csv')
#Y=data.pop('species')
species=dict(zip(list(data['species'].unique()), ([0, 1, 2])))

data['species'].replace(species,inplace=True)
Y=data.pop('species')
X=data.values
y_dummy=pd.get_dummies(Y)

```

```python
model=Sequential([ 
    Dense(32,activation='relu',input_dim=X.shape[1],name='layer1'),
    Dense(64,activation='relu',name='layer2'),
     Dense(64,activation='relu',name='layer3'),
    Dense(3,activation='softmax',name='out')

])
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['acc'])
```

```python
model.summary()
```

```python
model.fit(X,y_dummy,epochs=100)
```
